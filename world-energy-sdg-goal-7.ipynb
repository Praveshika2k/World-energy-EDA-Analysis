```python

import pandas as pd
```

```python
df=pd.read_csv("/kaggle/input/world-energy/World_Energy_By_Country_And_Region_1965_to_2023.csv")
df.head()
```

```python
# Convert numerical columns
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors="coerce")
df.head()
```

```python
# 1. Summary statistics
summary_stats = df.describe()
summary_stats
```

```python
# 2. Yearly total energy consumption
yearly_totals = df.iloc[:, 1:].sum()
yearly_totals
```

```python
# 3. Top 5 energy-consuming countries in 2023
top_5_2023 = df[['Country', '2023']].sort_values(by='2023', ascending=False).head(5)
top_5_2023
```

```python
# 4. Countries with the highest growth rate (1965-2023)
df['growth_rate'] = ((df['2023'] - df['1965']) / df['1965']) * 100
highest_growth_countries = df[['Country', 'growth_rate']].sort_values(by='growth_rate', ascending=False).head(5)
highest_growth_countries
```

```python
# 5. Year-wise correlation matrix
correlation_matrix = df.iloc[:, 1:].corr()
correlation_matrix
```

```python
# 6. Standard deviation for each country
std_dev_per_country = df.set_index('Country').std(axis=1)
std_dev_per_country
```

```python
# 8. Percentage contribution of top countries in 2023
total_2023 = df['2023'].sum()
df['percentage_2023'] = (df['2023'] / total_2023) * 100
top_percentage_countries = df[['Country', 'percentage_2023']].sort_values(by='percentage_2023', ascending=False).head(5)
top_percentage_countries
```

```python
# 9. Energy consumption trend for US & China
selected_countries = df[df['Country'].isin(['US', 'China'])].set_index('Country').T
selected_countries
```

```python
# 10. Outlier detection using IQR
Q1 = df.iloc[:, 1:].quantile(0.25)
Q3 = df.iloc[:, 1:].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df.iloc[:, 1:] < (Q1 - 1.5 * IQR)) | (df.iloc[:, 1:] > (Q3 + 1.5 * IQR))).sum().sum()
outliers
```

```python
#7. Mean, median and mode
df.iloc[:, 1:] = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')

def calculate_statistics(column):
    mean = column.mean()
    median = column.median()
    mode = column.mode().iloc[0] if not column.mode().empty else None
    return mean, median, mode

for col in df.columns[1:]:
    mean, median, mode = calculate_statistics(df[col])
    print(f"Column: {col}")
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}\n")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Line plot of total yearly energy consumption
plt.figure(figsize=(12,6))
plt.plot(yearly_totals.index, yearly_totals.values, marker='o')
plt.title('Total Energy Consumption Over the Years')
plt.xlabel('Year')
plt.ylabel('Energy Consumption')
plt.grid(True)
plt.show()
```

```python
# 2. Bar plot for top 5 energy-consuming countries in 2023
plt.figure(figsize=(8,5))
sns.barplot(data=top_5_2023, x='Country', y='2023')
plt.title('Top 5 Energy-Consuming Countries in 2023')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
# 3. Heatmap of correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False)
plt.title('Yearly Energy Consumption Correlation')
plt.show()
```

```python
# 4. Histogram of energy consumption distribution (2023)
plt.figure(figsize=(8,5))
sns.histplot(df['2023'].dropna(), bins=20, kde=True)
plt.title('Energy Consumption Distribution in 2023')
plt.xlabel('Energy Consumption')
plt.show()
```

```python
# 5. Boxplot of energy consumption for selected countries
plt.figure(figsize=(10,6))
selected_data = df[df['Country'].isin(['US', 'China', 'India', 'Russia', 'Japan'])].set_index('Country').T
sns.boxplot(data=selected_data)
plt.title('Energy Consumption Boxplot (Selected Countries)')
plt.xlabel('Country')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
# 6. Pie chart of top 5 countries’ percentage contribution in 2023
plt.figure(figsize=(8, 8))
plt.pie(top_percentage_countries['percentage_2023'], labels=top_percentage_countries['Country'], autopct='%1.1f%%')
plt.title('Top 5 Countries Contribution to Energy Consumption in 2023')
plt.show()
```

```python
# 7. Trend comparison of US vs China over the years
plt.figure(figsize=(12,6))
for country in ['US', 'China']:
    plt.plot(selected_countries.index, selected_countries[country], marker='o', label=country)
plt.title('Energy Consumption Trend: US vs China')
plt.xlabel('Year')
plt.ylabel('Energy Consumption')
plt.legend()
plt.show()
```

```python
# 8. Scatter plot comparing 2000 vs 2023 energy consumption
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['2000'], y=df['2023'])
plt.title('Energy Consumption in 2000 vs 2023')
plt.xlabel('Year 2000')
plt.ylabel('Year 2023')
plt.grid(True)
plt.show()
```

```python
#9.  Stream Graph: Energy Consumption Trends of Top Countries
df_top = df[df['Country'].isin(['US', 'China', 'India', 'Russia', 'Japan'])].set_index('Country').T
df_top.plot(kind='area', stacked=True, alpha=0.7, figsize=(12,6), colormap="coolwarm")
plt.title("Stream Graph: Energy Consumption Trends (Top 5 Countries)")
plt.xlabel("Year")
plt.ylabel("Energy Consumption")
plt.show()
```

```python
#10.  Hexbin Plot: Energy Consumption 2000 vs 2023
plt.figure(figsize=(8,6))
plt.hexbin(df['2000'], df['2023'], gridsize=30, cmap='Blues', mincnt=1)
plt.colorbar(label="Density of Countries")
plt.title("Hexbin Plot: Energy Consumption (2000 vs 2023)")
plt.xlabel("Year 2000 Energy Consumption")
plt.ylabel("Year 2023 Energy Consumption")
plt.show()
```
